```python
import pandas as pd
import json
```

```python
df = pd.read_csv('raw_data.csv')
df.head()
```

```python
data = df['dialogue']
data.head()
```

```python
data.to_list()
```

```python
dialogs = []
dialog_id = 1
for text in data:
    lines = text.splitlines()
    for line in lines:
        if ":" in line:
            person, sentence = line.split(':', 1)
            dialogs.append({'dialog_id': dialog_id, 
                            'person': person.strip(), 
                            'sentence': sentence.strip()})
    dialog_id += 1
print(json.dumps(dialogs, indent=2, ensure_ascii=False))
```

```python
with open("dialogs.json", "w", encoding="utf-8") as f:
    json.dump(dialogs, f, indent=2, ensure_ascii=False)
```
